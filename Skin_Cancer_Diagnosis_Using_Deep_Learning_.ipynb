#EfficientNetV2M + Neural Network

```python
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
import keras
from keras.optimizers import SGD
from keras.models import Sequential
from tensorflow.keras import layers
from tensorflow.keras.preprocessing import image_dataset_from_directory
from tensorflow.keras.applications import EfficientNetV2M
from tensorflow.keras.layers import Input, Add, Dense, Activation, , Dropout, BatchNormalization
from tensorflow.keras.models import Model, load_model
```

```python
#Prepare the data and define the input size and batch size

train_directory = "/kaggle/input/ham10000-noprocessing-diffsplit/HAM10000_noprocessing_diffsplit/train"
val_directory = "/kaggle/input/ham10000-noprocessing-diffsplit/HAM10000_noprocessing_diffsplit/val"
test_directory= "/kaggle/input/ham10000-noprocessing-diffsplit/HAM10000_noprocessing_diffsplit/test"
BATCH_SIZE = 16
IMG_SIZE = (224, 224)

train_dataset = image_dataset_from_directory(train_directory,
                                             shuffle=True,
                                             batch_size=BATCH_SIZE,
                                             image_size=IMG_SIZE,
                                             seed=42)
validation_dataset = image_dataset_from_directory(val_directory,
                                             shuffle=True,
                                             batch_size=BATCH_SIZE,
                                             image_size=IMG_SIZE,
                                             seed=42)

test_dataset =  image_dataset_from_directory(test_directory,
                                             shuffle=True,
                                             batch_size=BATCH_SIZE,
                                             image_size=IMG_SIZE,
                                             seed=42)

```

```python
batch_size = 16
AUTOTUNE = tf.data.AUTOTUNE

def prepare(ds, shuffle=False, augment=False):
  if shuffle:
    ds = ds.shuffle(1000)

  # Use data augmentation only on the training set.
  if augment:
    ds = ds.map(lambda x, y: (data_augmentation(x, training=True), y),
                num_parallel_calls=AUTOTUNE)

  # Use buffered prefetching on all datasets.
  return ds.prefetch(buffer_size=AUTOTUNE)
```

```python
train_dataset = prepare(train_dataset, shuffle=True)
validation_dataset = prepare(validation_dataset)
test_dataset = prepare(test_dataset)
```

```python
AUTOTUNE = tf.data.experimental.AUTOTUNE
train_dataset = train_dataset.prefetch(buffer_size=AUTOTUNE)
```

```python
# Load EfficientNetV2M model for feature extraction
efficientnet = EfficientNetV2M(weights='imagenet', include_top=False, input_shape=(224, 224, 3), pooling='avg')

# Freeze EfficientNet layers to prevent updates during feature extraction
efficientnet.trainable = False

# Define feature extraction model
feature_extractor = Model(inputs=efficientnet.input, outputs=efficientnet.output)

# Extract features from your dataset
def extract_features(dataset):
    features = []
    labels = []
    for images, lbls in dataset:
        feats = feature_extractor.predict(images)
        features.extend(feats)
        labels.extend(lbls)
    return tf.convert_to_tensor(features), tf.convert_to_tensor(labels)

# Extract features from train and validation datasets
train_features, train_labels = extract_features(train_dataset)
val_features, val_labels = extract_features(validation_dataset)
```

```python
#Extract features from the test dataset
test_features, test_labels = extract_features(test_dataset)
```

```python
# Define a classification head with dense, batch normalization and dropout layers
model = Sequential([
    Dense(128, activation='relu', input_shape=(1280,)),  # The output size of EfficientNetV2M
    BatchNormalization(),
    Dropout(0.2),
    Dense(128, activation='relu'),
    BatchNormalization(),
    Dropout(0.2),
    Dense(256, activation='relu'),
    BatchNormalization(),
    Dropout(0.3),
    Dense(256, activation='relu'),
    BatchNormalization(),
    Dropout(0.5),
    Dense(2, activation='softmax')
])

# Compile the model
model.compile(optimizer=tf.keras.optimizers.SGD(learning_rate=1e-3),
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Train the model on extracted features
history = model.fit(
    train_features, train_labels,
    validation_data=(val_features, val_labels),
    epochs=50,
    batch_size=16,
    callbacks=[
        tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=50, restore_best_weights=True),
        tf.keras.callbacks.ReduceLROnPlateau(
    monitor='val_loss', factor=0.5, patience=3, min_lr=1e-8
)
    ]
)
```

```python
#Plot training and validation accuracies
acc = [0.] + history.history['accuracy']
val_acc = [0.] + history.history['val_accuracy']

plt.figure(figsize=(8, 8))
plt.subplot(2, 1, 1)
plt.plot(acc, label='Training Accuracy')
plt.plot(val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.ylabel('Accuracy')
plt.ylim([min(plt.ylim()),1])
plt.title('Training and Validation Accuracy')

#Plot training and validation losses
loss = history.history['loss']
val_loss = history.history['val_loss']

plt.subplot(2, 1, 2)
plt.plot(loss, label='Training Loss')
plt.plot(val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.ylabel('Cross Entropy')
plt.title('Training and Validation Loss')
plt.xlabel('epoch')
plt.show()
```
